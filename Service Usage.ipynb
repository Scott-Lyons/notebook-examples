## Service Usage Map
This notebook is used to generate a visualisation of the usage of services by applications. The data is read in from a CSV file and then processed to create a network graph. This graph is then visualised using various plotting libraries to demonstration different examples.

### Import
Import data from a cleaned CSV file. An example CSV file is provided named `data_example_format.csv`.

```python
import pandas as pd
from pathlib import Path

# Create the exports directory ready for outputting any graphs
Path("./exports").mkdir(exist_ok=True)

#data = pd.read_csv("data_example_format.csv")
data = pd.read_csv("data_latest.csv")#print(data)
data.head(5)
```

### Arrange
Everything primarily uses networkx underneath so we start by building a networkx graph object populating the nodes and edges along with a few of the color attributes that will be used by different plotting tools

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

serviceNodes = data["Endpoints"].tolist()
G.add_nodes_from(serviceNodes)

colourMap = [];

applicationNodes = []

for key in data:
    if key != "Endpoints":
        applicationNodes.append(key)


G.add_nodes_from(applicationNodes);

serviceNodeColour = "blue"
applicationNodeColour = "orange"

# Assign the colours used by networkx
for node in serviceNodes:
    colourMap.append(serviceNodeColour);

for node in applicationNodes:
    colourMap.append(applicationNodeColour);

# Assign the colours to a colour attribute, these are used by other mapping libraries
for i, node in enumerate(G.nodes()):
    if node in serviceNodes:
        G.nodes[node]["colour"] = serviceNodeColour
    else:
        G.nodes[node]["colour"] = applicationNodeColour

# Create storage outside the networkx graph object for the edges
edge_raw = {"source":[], "target": [], "value": []}
edges = pd.DataFrame(edge_raw);

# Loop through our data headers and then service nodes to find any connections and add appropriate edges
for key in data:
    for index, value in enumerate(serviceNodes):
        if data[key][index] == 1.0:
            G.add_edge(key, value, weight=1)
            row = {"source": key, "target": value, "value": 1}
            edges = pd.concat([edges, pd.DataFrame([row])], ignore_index=True)
```

### networkx
This draws out a networkx chart using the nodes and edges generated in the previous cell. While this provides a lot of control over the positioning of elements, positioning is relative and can be a bit flaky. Leaving the position to its own devices (i.e. using the sprint_layout) can produce overlapping nodes.

```python
positions = {}
x = 0
everyOther = False
for node in serviceNodes:
    everyOther = not everyOther
    positions[node] = (x, 0 if everyOther else 1)
    x += 2

x = 0
for node in applicationNodes:
    everyOther = not everyOther
    positions[node] = (x, 4 if everyOther else 5)
    x += 1

#pos = nx.spring_layout(G, k=0.3*1/np.sqrt(len(G.nodes())), iterations=20)
pos = positions
#pos = nx.circular_layout(G)
#pos = nx.planar_layout(G)

plt.figure(figsize=(15, 15), dpi=150)
nx.draw(G, pos, with_labels=True, node_size=100, node_color=colourMap, font_size=3, font_weight='regular', arrows=False)
plt.title("Applications connected to Services")

plt.show()
#plt.savefig("./exports/graph.png", dpi=1000)
#plt.savefig("./exports/graph.pdf")
#plt.savefig("./exports/graph.svg")
```

### Holoviews - Datashader
Holoviews provides a few different chart types for visualising data. Specifically here we're using datashaders to connect associated connections together to try and reduce clutter making things easier to visualise. Because datashaders are recommended for larger datasets labels are not easy to apply. The chart is interactive and can be saved to a standalone html file.


```python
import holoviews as hv
from holoviews import opts
from holoviews.operation.datashader import bundle_graph

hv.extension('bokeh')

SG = (hv.Graph.from_networkx(G, nx.layout.fruchterman_reingold_layout, k=1)
      .opts(height=1000, width=1000, node_color='colour', cmap=[serviceNodeColour, applicationNodeColour]))
#SG
bundled = bundle_graph(SG)
bundled

#hv.save(bundled, './exports/interactive_chart.html')
```

### Holoviews - Sankey
Sanky is another layout of the Holoviews library and is useful for giving values to connections.

```python
import holoviews as hv

hv.extension('bokeh')

sankey = hv.Sankey(edges, label='Service Connections')
sankey.opts(label_position='left', edge_color='target', node_color='index', cmap='tab20')

#hv.save(sankey, './exports/sankey_diagram.html')
```

### Holoviews - Chords
Chords is a different layout of the Holoviews library and is useful for showing connections between nodes.

```python
import holoviews as hv

hv.extension('bokeh')

chords = hv.Chord(edges, label='Service Connections')
chords.opts(opts.Chord(cmap='Category20', height=800, labels='source', width=1000))

#hv.save(chords, './exports/chords_diagram.html')
```
