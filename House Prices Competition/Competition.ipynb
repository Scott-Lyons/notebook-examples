```python
import pandas as pd

house_data = pd.read_csv('./input/train.csv')
test_data = pd.read_csv('./input/test.csv')
sample_data = pd.read_csv('./input/sample_submission.csv')

display(house_data.describe())
display(test_data.describe())
display(sample_data.describe())
```

```python
features = ['LotArea', 'YearBuilt', 'FullBath', 'OverallQual', 'OverallCond', 'WoodDeckSF', 'GarageArea']

X = house_data[features]
y = house_data.SalePrice

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(X, y, random_state = 0)

house_model = RandomForestRegressor(random_state = 1)
house_model.fit(train_X, train_y)

preds = house_model.predict(test_X);

from sklearn.metrics import root_mean_squared_error
print(root_mean_squared_error(test_y, preds))

```

Apply our model to our data for compare against the sample submission

```python
testX = test_data[features]
predsy = house_model.predict(testX)

display(testX[testX.isnull().any(axis=1)])

output = pd.DataFrame({'Id': test_data.Id, 'SalePrice': predsy})
print("Test Data")
display(output.head(5))

print("Sample Data")
display(sample_data.head(5))
```
